## Exercise 2: Visualizing Complex Operations
1. Write a Python function to plot a complex number on the complex plane. Use it to plot `(3 + 4j)` and its conjugate.

```python
import matplotlib.pyplot as plt

def grafico_cplx(z):
    plt.figure(figsize=(6,6))
    ax = plt.gca()
    
    #grafica el numero complejo como punto.
    plt.plot([0,z.real],[0,z.imag],label=f"{z}(complejo)")

    #grafica el conjugado como punto.
    z_conjugado = z.conjugate()
    plt.plot([0,z_conjugado.real],[0,z_conjugado.imag],0,label=f"{z_conjugado}(conjugado)")

    #ejes
    plt.axhline(0, color='black',linewidth=1)
    plt.axvline(0, color='black',linewidth=1)
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)

    # Etiquetas y título
    plt.xlabel('Parte Real')
    plt.ylabel('Parte Imaginaria')
    plt.title('Número Complejo y su Conjugado')
    plt.grid(True)

    plt.legend()

    # Mostrar la gráfica
    plt.show()

    #llamo a la funcion grafico_cplx, para graficar3+4i
grafico_cplx(3+4j)
```

2. Extend the function to show addition and multiplication of two complex numbers graphically.

```python
import matplotlib.pyplot as plt

def graficar_complejos_operaciones(z1, z2):
    # Crear la figura y los ejes
    plt.figure(figsize=(8, 8))
    ax = plt.gca()

    # Trazar el número complejo z1 (como un punto)
    plt.plot([0,z1.real], [0,z1.imag], 'bo', label=f'{z1} (Complejo 1)')
    # Trazar el número complejo z2 (como un punto)
    plt.plot(z2.real, z2.imag, 'go',label=f'{z2} (Complejo 2)')

    # Trazar los conjugados
    z1_conjugado = z1.conjugate()
    z2_conjugado = z2.conjugate()
    plt.plot([0,z1_conjugado.real], [0,z1_conjugado.imag], 'ro', label=f'{z1_conjugado} (Conjugado 1)')
    plt.plot([0,z2_conjugado.real], [0,z2_conjugado.imag], 'yo', label=f'{z2_conjugado} (Conjugado 2)')

    # Sumar complejos
    suma = z1 + z2
    plt.plot([0,suma.real], [0,suma.imag],color='purple', label=f'Suma: {suma} (z1 + z2)')

    # Multiplicar complejos
    multiplicacion = z1 * z2
    plt.plot([0,multiplicacion.real], [0,multiplicacion.imag],color='orange', label=f'Multiplicación: {multiplicacion} (z1 * z2)')

    # Etiquetas de los ejes
    plt.axhline(0, color='black',linewidth=1)
    plt.axvline(0, color='black',linewidth=1)
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)

    # Etiquetas y título
    plt.xlabel('Parte Real')
    plt.ylabel('Parte Imaginaria')
    plt.title('Operaciones con Números Complejos')
    plt.grid(True)

    # Mostrar la leyenda
    plt.legend()

    # Mostrar la gráfica
    plt.show()

# Llamada a la función con dos números complejos
graficar_complejos_operaciones(3 + 4j, 1 + 2j)
```

## Exercise 3: Exploring the Mandelbrot Set
1. Modify the provided Mandelbrot set code to change its zoom level and center point. Observe how the fractal pattern changes.
2. Experiment with different values of `max_iter` and observe the effect on the fractal's detail and computation time.

```python
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = 0
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter
#zoom
zoom=4.5
center_x,center_y=-0.2,1

# Create a grid of complex numbers
pixels = 800
max_iter = 1000

#Rangos zoom
x_min,x_max=center_x-(2/zoom),center_x+(2/zoom)
y_min,y_max=center_y-(2/zoom),center_y+(2/zoom)

x = np.linspace(x_min, x_max, pixels)
y = np.linspace(y_min, y_max, pixels)
xx, yy = np.meshgrid(x, y)
c = xx + yy*1j

# Compute Mandelbrot set
mandelbrot_set = np.array([mandelbrot(ci, max_iter) for ci in c.ravel()]).reshape(c.shape)

# Plot
plt.imshow(mandelbrot_set, extent=[-2, 1, -2, 1.5])
plt.colorbar()
plt.title("Mandelbrot Set")
plt.show()
```

El comportamiento que  se evidencia en el fractal es, que al momento de compilar entre mayor sea el max_iter, se demorara un poco mas y en cuanto a la imagen, se va haciendo mas redonda y va aumentando la cantidad de patrones.
